Реальные скрипты колл-центра. После фильтрации по словам cancel/churn/discount дает реальные сценарии возвращаемости клиентов

```python
df = load_dataset("bitext/Bitext-customer-support-llm-chatbot-training-dataset")
df1 = df["train"].to_pandas()
intent = ["track_order", "delivery_options", "contact_human_agent","get_invoice",
          "check_invoice", "registration_problems", "payment_issue", "check_payment_methods"]
dffinal = df1[df1["intent"].isin(intent)].copy()
dffinal.to_csv("callcenter_scripts.csv", index=False)
print(dffinal.shape)
dffinal.head(3)
```
